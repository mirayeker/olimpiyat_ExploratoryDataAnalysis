```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veri=pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/exploratory_data_analysis/olimpiyatlar_temizlenmis.csv")
veri.head()
```

# Saçılım Grafiği
- saçılım grafiği, iki farklı değişken arasındaki ilişkiyi elirlemek için kullanılır

```python
def sacilimGrafikCiz():
  sns.scatterplot(x="boy",y="kilo",data=veri)
  plt.title("Boy ve Ağırlık İlişkisi ")
  plt.show()
sacilimGrafikCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafikCiz()
```

```python
sns.scatterplot(x="boy",y="kilo",hue ="madalya",data=veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy",y="kilo",hue ="cinsiyet",style="madalya",data=veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
# dört boyutlu görselleştirme 
```

```python
sns.scatterplot(x="boy",y="kilo",hue ="cinsiyet",style="madalya",size="yas",sizes=(15,200),data=veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
# beş boyutlu görselleştirme 
```

## Saçılım Grafiği - Doğrusal Regresyon

```python
sns.regplot(x="boy",y="kilo",data=veri,marker="+",scatter_kws={"alpha":0.2})
plt.title("Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy",y="kilo",hue="madalya", data=veri, palette="Set1")
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

# Çizgi Grafiği 

```python
sns.lineplot(x="boy",y="kilo",data=veri)
plt.title("Boy ve Kilo")
plt.show()
```

## Çizgi Grafiği - Kategorik

```python
sns.lineplot(x="boy",y="kilo",hue="cinsiyet",data=veri)
plt.title("Cinsiyete Göre Boy ve Ağırlık İlişkileri")
plt.show()
```

# Histogram Grafiği
- Histogram, gruplandırılmış bir veri dağılımın sütun grafiğiyle gösterimidir.
- Tekrarlı sayılardan oluşan verilerin, uygulanan işlemlemlerden önce tabloya, tablodan yararlanarak grafiğe aktarılması, yani veri gruplarının grafiğinin dikdörtgen sütunları halinde gösterilmesidir.

```python
sns.displot(veri,x="kilo")
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

## Histogram - Kategorik

```python
sns.displot(veri,x="kilo",hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete Göre Ayrıştırılmış Ağırlık histogramı")
plt.show()
```

```python
sns.displot(veri,x="kilo",col="cinsiyet",multiple="dodge")

plt.show()
```

## İki Boyutlu Histogram

```python
sns.displot(veri,x="kilo",y="boy",kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("boy")
plt.title("CAğırlık - Boy histogramı")
plt.show()
```

```python
sns.displot(veri,x="kilo",y="boy",hue="cinsiyet" , kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("boy")
plt.title("Cinsiyete göre Ayrıştırılmış Ağırlık - Boy histogramı")
plt.show()
```

```python
sns.kdeplot(data=veri,x="kilo",y="boy",hue="cinsiyet",fill=True)
plt.xlabel("Ağırlık")
plt.ylabel("boy")
plt.title("Cinsiyete göre Ayrıştırılmış Ağırlık - Boy histogramı")
plt.show()
```

# Çubuk Grafiği 
- Çubuk grafik, kategorik verileri temsil ettikleri değerlerle orantılı yükseklik veya uzunluktaki dikdörtgen çubuklarla sunan bir grafiktir

```python
sns.barplot(x="madalya",y="boy",data=veri)
plt.title("Boylara göre madalya grafikleri")
plt.show()
```

```python
sns.barplot(x="madalya",y="boy",hue="cinsiyet",data=veri)
plt.title("Cinsiyete göre madalya grafikleri")
plt.show()
```

```python
sns.barplot(x="madalya",y="yas",hue="cinsiyet",data=veri)
plt.title("Cinsiyete göre madalya grafikleri")
plt.show()
```

```python
sns.catplot(x="madalya",y="yas", hue="cinsiyet",col="sezon",data=veri,kind="bar",height=4,aspect=0.7)
```

# Kutu Grafiği
- istatistik biliminde kutu grafiği bir betimsel istatistik ve istatiksel grafik aleti olup niceliksel verileri görsel şekilde özetlemek için kullanılır.

```python
sns.boxplot(x="sezon",y="boy",data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon",y="boy",hue="cinsiyet",data=veri,palette="Set2")
plt.show()
```

```python
veri_gecici=veri.loc[:,["yas","kilo","boy"]]
sns.boxplot(data=veri_gecici,orient="h",palette="Set2")
plt.show()
```

```python
sns.catplot(x="sezon",y="boy",col="madalya",data=veri,kind="box",height=4,aspect=0.7)
```

# Isı Haritası

```python
sns.heatmap(veri.corr(),annot=True,linewidths=0.5,fmt=".2f")
plt.show()
```

# Keman Grafiği 
- Keman grafiği, sayısal verileri çizme yöntemidir.
- Kutu grafiğine benzer yorumlanır


```python
sns.violinplot(x="sezon",y="boy",data=veri)
plt.show()
```

```python
sns.violinplot(x="sezon",y="boy",hue="cinsiyet",data=veri)
plt.show()
```

```python
sns.catplot(x="sezon",y="boy",hue="cinsiyet",col="madalya",data=veri,kind="violin",split=True,height=4,aspect=0.7)
```

# Ortaklık Grafiği
- Farklı özellikler arasındaki ilişkiyi farklı görselleştirme yöntemleri kullanarak incelemeye yarayan bir grafiktir.

```python
sns.jointplot(data=veri, x="kilo",y="boy",hue="sezon", kind="kde")
plt.show()
```

```python
g= sns.JointGrid(data=veri,x="kilo",y="boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

# Çift/Eş Grafiği 

```python
sns.pairplot(veri)
plt.show()
```

```python
g=sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
